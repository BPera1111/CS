```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 100  # Hz
fs = 1000  # Hz
t_i = 0.0  # s
t_f = 0.5  # s
```

```python
Ts = 1/fs
n_i = int(t_i // Ts)
n_f = int(t_f // Ts)
time = np.array([n*Ts for n in range(0, n_f)])
signal = np.sin(2*np.pi*f*time)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(time, signal)
ax.set_xlabel('time (s)')
ax.set_ylabel('signal')
```
